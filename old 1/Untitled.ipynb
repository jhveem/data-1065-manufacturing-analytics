```python
for index, row in df_motors.iterrows():
    for i in range(0, 4):
        motor_number = i + 1
        current_value = row[i]
        motor_column = df_motors.iloc[i, 0: index + 1]
        motor_mean = motor_column.mean()
        if current_value > motor_mean * 1.25:
            print('\t', motor_number, motor_mean, current_value)
```

```python
df_motors['mean'] = df_motors.mean(axis=1)
```

```python
for index, row in df_motors.iterrows():
    mean = row[4]
    for i in range(0, 4):
        motor_number = i + 1
        motor_amps = row[i]
        if motor_amps > mean * 1.5:
            print(motor_number)
```
